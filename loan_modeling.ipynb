```python
# ==========================================
# EPIC 4: MODEL BUILDING & EVALUATION
# ==========================================
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Load the preprocessed data arrays
data = joblib.load('models/preprocessed_loan_data.pkl')
X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']

print("Data loaded successfully! Ready to train models.\n")

# 2. Initialize the specific models required by your Epic 4 spec
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN Model": KNeighborsClassifier(),
    "XGBoost Model": xgb.XGBClassifier(random_state=42, eval_metric='logloss')
}

# 3. Train and print performance metrics
for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, predictions)
    print(f"=== {name} ===")
    print(f"Accuracy Score: {accuracy * 100:.2f}%")
    print(classification_report(y_test, predictions))
    print("-" * 40)
```

```python
# ==========================================
# EPIC 4: FINAL EVALUATION & MODEL SAVING
# ==========================================
import joblib

# Let's find your absolute best model dynamically
best_model_name = None
best_accuracy = 0
best_model_object = None

for name, model in models.items():
    predictions = model.predict(X_test)
    acc = accuracy_score(y_test, predictions)
    if acc > best_accuracy:
        best_accuracy = acc
        best_model_name = name
        best_model_object = model

print(f"🏆 THE WINNER IS: {best_model_name} with an Accuracy of {best_accuracy * 100:.2f}%\n")

# Bundle the winning model with the scaler we used during preprocessing
# This is crucial so the Flask App scales live user inputs the exact same way!
model_artifacts = {
    'model': best_model_object,
    'scaler': data['scaler'] # Using the scaler carried over from Epic 3
}

# Save it to disk
joblib.dump(model_artifacts, 'models/best_loan_model.pkl')
print("💾 Success! Your winning model and scaler have been saved together as 'best_loan_model.pkl'")
```
